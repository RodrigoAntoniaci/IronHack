# Shark Attacks - Project of Data Cleaning/Manipulation

### The archive `attacks.csv` was obtained in [Kaggle](https://www.kaggle.com/teajay/global-shark-attacks#attacks.csv). 

# Objective:

### Identifying if near the release date of the movie Jaws, there was some anomally against other years attacks




# Importing libs

```python
import pandas as pd
import re
import numpy as np
import seaborn as sns
```

# Read data from the archive `'attacks.csv'`, utilized enconding `latin_1`

```python
df = pd.read_csv('./attacks.csv', sep = ',', encoding='latin_1')
```

# Getting to know the dataframe

```python
df.info()
```

# Verification of how many collumns has  `missing` values and the respective percentage

```python
df.isnull().mean().round(5)
```

# Removing useless columns `Unnamed: 22`  and  `Unnamed:23`

```python
df[(df['Unnamed: 23'].isnull())==False]
```

```python
df[(df['Unnamed: 22'].isnull())==False]
```

```python
df = df.drop(columns=['Unnamed: 22','Unnamed: 23'])
```

# Removing empty lines of the DataFrame

```python
df = df.dropna(axis=0,how='all', thresh=3)
df.shape
```

# Adjusting columns names

```python
new_columns = []
for col in df.columns:
    new_columns.append(col.lower().strip())

df.columns = new_columns
```

```python
df.columns
```

# Picking only the interesting columns

```python
df = df.drop(columns=['case number','name','sex','age','time',
                      'investigator or source','location','pdf',
                      'href formula','href','original order',
                      'case number.1','case number.2','country'])
```

# Replacing empty fields for `Not_classified`

```python
df = df.mask(df.isnull(),'Not_classified')
```

# Adjusting <b> YEAR </b> column:

```python
df.loc[df['year'] == 'Not_classified']
```

```python
df['year'][187] = 2017
df['year'][6079] = 1836
```

# Dropping column <b> DATE </b>

```python
df = df.drop(columns=['date'])
```

# Dropping attacks before 1900

```python
df = df.drop(df[df.year < 1900].index)
```

# Adjusting values of the <b> TYPE </b> column:

```python
df['type'].unique()
```

```python
df['type'] = df['type'].mask((df['type'] == 'Boatomg') |
                             (df['type'] == 'Boating'),'Boat')

df['type'] = df['type'].mask(df['type'] == 'Invalid','Not_classified')
```

# Adjusting values of the <b> AREA </b> column:

```python
df['area'] = df['area'].apply((lambda x: x.capitalize().strip()))
df['area'] = df['area'].mask(df['area'] == '10ºs, 142ºe','Guam')
df['area'] = df['area'].mask(df['area'] == '19s, 178?','Wake Island')
df['area'] = df['area'].mask(df['area'] == '9.35n 79.35w','Dhanushkodi')
df['area'] = df['area'].mask(df['area'] == '04.05n-13.23w','Liberia')
df['area'] = df['area'].mask(df['area'] == '22ºn, 88ºe','Calcuta')
df['area'] = df['area'].mask(df['area'] == '35º39 : 165º8\'','Midway Atoll')
df['area'] = df['area'].mask(df['area'] == '18s / 50e','Antananaviro')
df['area'] = df['area'].mask(df['area'] == '19s, 178?e','Kadavu Island')
```

# Using column <b> INJURY </b> to update <b> FATAL (Y/N)</b> column:

```python
df['injury'].unique()
```

```python
cond_fatal = df['injury'].str.match('fatal',flags=re.IGNORECASE)
df['fatal (y/n)'][cond_fatal] = 'Y'
```

# Dropping column <b> INJURY </b>

```python
df = df.drop(columns=['injury'])
```

# Adjusting values of the <b> FATAL (Y/N)</b> column:

```python
df['fatal (y/n)'].unique()
```

```python
df['fatal (y/n)'] = df['fatal (y/n)'].mask((df['fatal (y/n)'] == 'N ') | 
                                           (df['fatal (y/n)'] == ' N') | 
                                           (df['fatal (y/n)'] == 'M') 
                                           ,'N')

df['fatal (y/n)'] = df['fatal (y/n)'].mask((df['fatal (y/n)'] == 'UNKNOWN') |
                                           (df['fatal (y/n)'] == '2017')
                                           ,'Not_classified')

df['fatal (y/n)'] = df['fatal (y/n)'].mask(df['fatal (y/n)'] == 'y','Y')
```

# Adjusting type of the column Year

```python
df['year'] = pd.to_numeric(df['year'],downcast='integer')
```

# Plotting histogram: Count of attacks since <b> 1900 </b>

```python
attacks_per_year = df.groupby(by=['year'],as_index=False,)[['fatal (y/n)']].count()
hist_attacks_per_year = sns.lineplot(data=attacks_per_year,x='year',y='fatal (y/n)')
```

# Plotting categorical histogram: Count of attacks since <b> 1900 </b>

```python
cat_hist_attacks_per_year = df.groupby(by=['year','fatal (y/n)'],as_index=False)[['type']].count()
sns.lineplot(x="year", y="type", hue="fatal (y/n)", data=cat_hist_attacks_per_year)
```

# Plotting histogram: Count of attacks between <b> 1960 - 1990 </b>

```python
attacks_per_year = df.loc[df['year'].between(1960,1990)].groupby(by=['year'],as_index=False,)[['fatal (y/n)']].count()
hist_attacks_per_year = sns.lineplot(data=attacks_per_year,x='year',y='fatal (y/n)')
```

# Plotting categorical histogram: Count of attacks between <b> 1960 - 1990 </b>

```python
cat_hist_attacks_per_year = df.loc[df['year'].between(1960,1990)].groupby(by=['year','fatal (y/n)'],as_index=False)[['type']].count()
sns.lineplot(x="year", y="type", hue="fatal (y/n)", data=cat_hist_attacks_per_year)
```

```python
attacks_1975 = df.loc[df['year'] == 1975].groupby(by=['year','fatal (y/n)'],as_index=False)[['type']].count()
sns.barplot(x='fatal (y/n)', y='type', data= attacks_1975)
```

```python
attacks_area = df.loc[df['year'] == 1975].groupby(by=['area'],as_index=False)[['type']].count().sort_values(by='type', ascending=False)
sns.barplot(x='type', y='area', data= attacks_area.head(5))
```

```python
attacks_activity = df.loc[df['year'] == 1975].groupby(by=['activity'],as_index=False)[['type']].count().sort_values(by='type', ascending=False)
sns.barplot(x='type', y='activity', data= attacks_activity.head(5))
```

```python
attacks_species = df.loc[df['year'] == 1975].groupby(by=['species'],as_index=False)[['type']].count().sort_values(by='type', ascending=False)
sns.barplot(x='type', y='species', data= attacks_species.head(5))
```
